```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
input_image = cv2.imread('Task1.jpg')
input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
inverted_image = 255 - input_image
gamma = 3
gamma_image = np.power(inverted_image / 255.0, gamma) * 255.0
gamma_image = gamma_image.astype(np.uint8)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(input_image)
plt.title('Input Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(gamma_image)
plt.title('Night Version Image')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
input_image = cv2.imread('Task2.jpg')
input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
denoised_image = cv2.medianBlur(input_image, 5)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(input_image)
plt.title('Input Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(denoised_image)
plt.title('Denoised Image')
plt.axis('off')
plt.show()

# The filtering method I have used is Median Filtering.
# This method replaces each pixel with the median value of the surrounding pixels, 
# which effectively removes noise while preserving the edges and details.
# Median filtering removes the noise efficiently without compromising the image’s sharpness, making it the most suitable choice.
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

blurred_image = cv2.imread('Task3.jpg')
blurred_image = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB)

gaussian_blurred = cv2.GaussianBlur(blurred_image, (9, 9), 10.0)
mask = blurred_image - gaussian_blurred

sharpness_factor = 5 
sharpened_image = blurred_image + sharpness_factor * mask
sharpened_image = np.clip(sharpened_image, 0, 255).astype(np.uint8)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(blurred_image)
plt.title('Blurred Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(sharpened_image)
plt.title('Sharpened Image')
plt.axis('off')
plt.show()
# The method I have used is Unsharp Masking.
# Unsharp masking provides a more natural enhancement with noticeable but not overly strong edge enhancement.
# unsharp masking is preferred over the Laplacean filter for image sharpening 
# due to its finer control over sharpening intensity, more natural results, better noise management.
```
